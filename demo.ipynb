# Demo for Neural Network

```python
import numpy as np
import matplotlib.pyplot as plt
from network import Network
from test_utils import test_accuracy, show_one, FASHION_MAP, COUNTER

# Load the Dataset
data = np.load('datasets/fashion_mnist.npz')
# data = np.load('datasets/mnist.npz')
x_train = data["x_train"] / 255 # NOTE normalizing the data here
y_train = data["y_train"]
x_test  = data["x_test"] / 255
y_test  = data["y_test"]

# Initialization and Training
net = Network(shape=[784,256,128,10])
net.descend_stoc(x_train, y_train, max_steps=1000)
print(f'accuracy: {test_accuracy(net, x_test, y_test)*100:.1f}%')
```

```python
# Seeing the Results
X_IDX = next(COUNTER)
prediction = FASHION_MAP[np.argmax(net.f_prop(x_input=x_test[X_IDX], output_only=True))]
# prediction = np.argmax(net.f_prop(x_input=x_test[X_IDX], output_only=True))

print(f'network prediction: {prediction}')
show_one(x_test[X_IDX])
```
